```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Base URL
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
```

```python
# List to store book data
books = []
```

```python
# Loop through all pages (1 to 50)
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
print(soup.title.text)
```

```python
# Find all book containers
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    book_containers = soup.find_all('article', class_='product_pod')

print(book_containers)
```

```python
# Make Containers
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    book_containers = soup.find_all('article', class_='product_pod')

    for book in book_containers:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        availability = book.find('p', class_='instock availability').text.strip()
        link = book.h3.a['href']  # Get the relative link to the book
        full_link = f"https://books.toscrape.com/catalogue/{link}"  # Create the full URL
        
```

```python
# Append book data to the list
for page in range(1, 51):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

  
    book_containers = soup.find_all('article', class_='product_pod')

    for book in book_containers:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text.strip()
        price = price.replace('Â', '')  # Remove unwanted character
        availability = book.find('p', class_='instock availability').text.strip()
        link = book.h3.a['href']  # Get the relative link to the book
        full_link = f"https://books.toscrape.com/catalogue/{link}"  # Create the full URL
        
       

        
        books.append({
            'Title': title,
            'Price': price,
            'Availability': availability,
            'Link': full_link # Add the link to the book
        })
```

```python
# Create a DataFrame 
Books = pd.DataFrame(books)
print(Books)
```

```python
# To see all column names
print(Books.columns)
```

```python
# To print a specific column
print(Books['Title']) 
```

```python

# First remove £ symbol using str.replace()
Books['Price'] = Books['Price'].str.replace('£', '')

# Then convert to float to handle decimal values
Books['Price'] = Books['Price'].astype(float)

# If you want integers (will truncate decimals):
Books['Price'] = Books['Price'].astype(int)

# Or to round to nearest integer:
#Books['Price'] = Books['Price'].round().astype(int)

# Print the cleaned prices
print(Books['Price'])
```
